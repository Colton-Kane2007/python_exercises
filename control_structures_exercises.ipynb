```python
# Prompt the user for an odd number between 1 and 50. Use a loop and a break statement to
# continue prompting the user if they enter invalid input. 
# (Hint: use the isdigit method on strings to determine this). 
# Use a loop and the continue statement to output all the odd numbers between 1 and 50, except for the number the user entered.
while True:
    i = input("Please enter an odd number between 1 and 50: ")
    if not i.isdigit():
        print("Invalid input. Please enter a valid number.")
        continue
    num = int(i)
    if num < 1 or num > 50 or num % 2 == 0:
        print("Invalid input. Please enter an odd number between 1 and 50.")
        continue
    break
print("Number to skip is: ", num)
for x in range(1, 51, 2):
    if x == num:
        continue
    print("Here is an odd number: ", x)

```

```python
# Prompt the user to enter an integer
# Display a table of squares and cubes from 1 to the value entered
# Ask if the user wants to continue
# Assume that the user will enter valid data
# Only continue if the user agrees to

def print_table(start, end):
    print("Number\tSquare\tCube")
    for i in range(start, end+1):
        square = i ** 2
        cube = i ** 3
        print(f"{i}\t{square}\t{cube}")

while True:
    num = int(input("What number would you like to go up to? "))

    print_table(1, num)

    choice = input("Do you want to continue? (yes/no): ")
    if choice.lower() != "yes":
        break
```

```python
# Prompt the user for a numerical grade from 0 to 100
# Display the corresponding letter grade
# Prompt the user to continue
# Assume that the user will enter valid integers for the grades
# The application should only continue if the user agrees to
# Grade Ranges:
# A : 100 - 88
# B : 87 - 80
# C : 79 - 67
# D : 66 - 60
# F : 59 - 0
def get_letter_grade(grade):
    if grade >= 88:
        return 'A'
    elif grade >= 80:
        return 'B'
    elif grade >= 67:
        return 'C'
    elif grade >= 60:
        return 'D'
    else:
        return 'F'

while True:
    grade = int(input("Enter a numerical grade from 0 to 100: "))
    letter_grade = get_letter_grade(grade)
    print("Letter grade:", letter_grade)

    choice = input("Do you want to continue? (yes/no): ")
    if choice != 'yes':
        break
```
